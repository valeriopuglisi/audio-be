```python
import matplotlib.pyplot as plt
from speechbrain.dataio.dataio import read_audio
from IPython.display import Audio
import os
import torchaudio
from speechbrain.dataio.preprocess import AudioNormalizer
from speechbrain.pretrained import EncoderDecoderASR
import os
```

```python
#LOAD AUDIO FILE
    datset_path = 'media'
    src_audio_name = 'common_voice_it_17415771.wav'
    src = os.path.join(datset_path, src_audio_name)
```

```python

signal, sr = torchaudio.load(src, channels_first = False)
normalizer = AudioNormalizer(sample_rate=16000)
normalized = normalizer(signal, sr)
Audio(normalized, rate=16000)
```

```python
clean = read_audio(src).squeeze()
Audio(clean, rate=16000)
```

```python
asr_model = EncoderDecoderASR.from_hparams(source="speechbrain/asr-wav2vec2-commonvoice-it", savedir="pretrained_models/asr-wav2vec2-commonvoice-it")
print(asr_model.transcribe_file(normalized))
```
